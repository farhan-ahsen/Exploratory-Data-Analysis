```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Replacing blank values in TotalCharges as 0 as Tenure of cust is 0

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

# Changing Senior Citizen to Boolean

```python
def conv(value):
    if value == 1:
        return "yes"
    else :
        return "no"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot( x = 'Churn' , data = df)

ax.bar_label(ax.containers[0])
plt.show()
```

# Grouping chrun values and displaying in a pie chart

```python
plt.figure(figsize = (4,4))
gb= df.groupby('Churn').agg({'Churn' : 'count'})
gb
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of churned customers")
plt.show()
```

# Customers Churned by Gender

```python
plt.figure(figsize = (4,4))
sns.countplot(x = 'gender', data = df, hue = 'Churn')
plt.title("Churned by Gender")
plt.show()
```

```python
plt.figure(figsize = (9,4))
sns.histplot( x = "tenure" , data = df, bins = 72, hue = 'Churn')
plt.show()
```

```python
df.columns.values
```

```python
# Columns to plot
columns_to_plot = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Subplot setup
n_cols = 3
n_rows = -(-len(columns_to_plot) // n_cols)  # Calculate rows required
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each countplot
for i, column in enumerate(columns_to_plot):
    sns.countplot(data=df, x=column, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Countplot of {column}')
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel(column)

# Hide unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Customers without certain services or with "No" features (e.g., OnlineSecurity or TechSupport) tend to have higher churn rates. Fiber optic internet users also show increased churn compared to DSL users.

```python
plt.figure(figsize = (6,4))
ax = sns.countplot  (x = "PaymentMethod", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```
